#### Detetcting Outliers

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

outliers = detect_outliers_iqr(df['AMT_INCOME_TOTAL'])
print(outliers.sum())
```

```python
# Detect outliers for each column
outliers_iqr_col1 = detect_outliers_iqr(df['AMT_INCOME_TOTAL'])
outliers_iqr_col2 = detect_outliers_iqr(df['CNT_CHILDREN'])
outliers_iqr_col3 = detect_outliers_iqr(df['CNT_FAM_MEMBERS'])
outliers_iqr_col4 = detect_outliers_iqr(df['DAYS_EMPLOYED'])

# Combine outlier flags for all columns
combined_outliers_iqr = outliers_iqr_col1 | outliers_iqr_col2 | outliers_iqr_col3 | outliers_iqr_col4

# Remove rows containing any outliers
df_cleaned_iqr = df[~combined_outliers_iqr]
```

#### Feature Engineering

```python
data['AGE'] = -data['DAYS_BIRTH'] // 365
```

```python
data['EMPLOYMENT_LENGTH'] = -data['DAYS_EMPLOYED'] // 365
```

```python
# Income category
bins = [0, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, float('inf')]
labels = ['Low', 'Low-Medium', 'Medium', 'Medium-High', 'High', 'Very High', 'Ultra High', 'Hyper High', 'Mega High', 'Giga High', 'Tera High']
data['INCOME_CATEGORY'] = pd.cut(data['AMT_INCOME_TOTAL'], bins=bins, labels=labels)
```

```python
# Family size calculation
data['FAMILY_SIZE'] = data['CNT_CHILDREN'] + data['CNT_FAM_MEMBERS']
```

```python
# Income per family member calculation
data['INCOME_PER_MEMBER'] = data['AMT_INCOME_TOTAL'] / data['FAMILY_SIZE']
```

```python
# Has phone binary feature
data['HAS_PHONE'] = data[['FLAG_WORK_PHONE', 'FLAG_PHONE', 'FLAG_MOBIL']].sum(axis=1) > 0
data['HAS_PHONE'] = data['HAS_PHONE'].astype(int)
```

```python
data['MARITAL_LIVING_INTERACTION'] = data['NAME_FAMILY_STATUS'] + '_' + data['NAME_HOUSING_TYPE']
```

```python
data['EDUCATION_INCOME_INTERACTION'] = data['NAME_EDUCATION_TYPE'] + '_' + data['AMT_INCOME_TOTAL'].astype(str)
```
